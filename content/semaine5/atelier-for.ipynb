### Exercice 6 : Trouver les nombres pairs
Écris un programme qui utilise une boucle `for` pour afficher tous les nombres pairs de 1 à 20.

```python
for i in range(1, 21):
    if i % 2 == 0:
        print(i)
```

### Exercice 8 : Générer une table de multiplication
Écris un programme qui utilise une boucle `for` pour générer et afficher la table de multiplication d'un nombre donné.

```python
def table_multiplication(n):
    for i in range(1, 11):
        print(f"{n} x {i} = {n * i}")

nombre = 7
table_multiplication(nombre)
```

### Exercice 9 : Calculer la moyenne des éléments d'une liste
Écris un programme qui utilise une boucle `for` pour calculer la moyenne des éléments d'une liste de nombres.

```python
def calculer_moyenne(liste):
    somme = 0
    for nombre in liste:
        somme += nombre
    return somme / len(liste)

ma_liste = [10, 20, 30, 40, 50]
print(f"La moyenne des éléments de la liste est {calculer_moyenne(ma_liste)}")
```

### Exercice 10 : Afficher une pyramide d'étoiles
Écris un programme qui utilise une boucle `for` pour afficher une pyramide d'étoiles de hauteur donnée.

```python
def pyramide_etoiles(hauteur):
    for i in range(1, hauteur + 1):
        print('*' * i)

hauteur = 5
pyramide_etoiles(hauteur)
```

### Exercice 10 : Inverser chaine

```python
def inverser_chaine(chaine):
    chaine_inversee = ""
    for caractere in chaine:
        chaine_inversee = caractere + chaine_inversee
    return chaine_inversee

# Exemple d'utilisation
texte = "Bonjour"
print(f"La chaîne inversée de '{texte}' est '{inverser_chaine(texte)}'")
```

Ce programme définit une fonction `inverser_chaine` qui prend une chaîne de caractères `chaine` en entrée et utilise une boucle `for` pour inverser cette chaîne. Ensuite, il affiche la chaîne inversée de "Bonjour" comme exemple.
