Un exemple concret d’utilisation de la régression linéaire en Python pour prédire les prix des maisons en fonction de leur superficie.

Supposons que nous avons un jeu de données avec la superficie des maisons (en mètres carrés) et leurs prix (en milliers de dollars).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Création des données
data = {
    'Superficie': [50, 60, 80, 100, 120, 150],
    'Prix': [150, 180, 240, 300, 360, 450]
}
df = pd.DataFrame(data)
X = df[['Superficie']]
y = df['Prix']

# Création du modèle de régression linéaire
modele = LinearRegression()

# Entraînement du modèle
modele.fit(X, y)

# Prédiction des prix pour les superficies données
y_pred = modele.predict(X)

# Tracé des points de données
plt.scatter(X, y, color='blue', label='Données réelles')

# Tracé de la ligne de régression
plt.plot(X, y_pred, color='red', label='Ligne de régression')

# Ajout de l'équation de la droite de régression
a = modele.coef_[0]
b = modele.intercept_
equation = f'y = {a:.2f}x + {b:.2f}'
plt.text(1, 4.5, equation, fontsize=12, color='black')

# Ajout des labels et de la légende
plt.xlabel('Superficie (m²)')
plt.ylabel('Prix (en milliers de dollars)')
plt.title('Régression Linéaire : Prédiction des prix des maisons')
plt.legend()

# Affichage du graphique
plt.show()

```
