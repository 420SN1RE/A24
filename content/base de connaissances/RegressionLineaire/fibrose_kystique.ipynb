Voici les étapes en Python pour générer une courbe étalon (ou courbe standard) afin de quantifier la trypsine en fonction de l'absorbance mesurée.

### Étape 1 : Importation des bibliothèques

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Étape 2 : Préparation des données
Supposons que nous avons les valeurs d'absorbance pour la solution mère et les 7 dilutions. Voici un exemple de données :

```python
# Tableau contenant les concentrations en trypsine (μg/ml)
concentrations = np.array([100, 50, 25, 12.5, 6.25, 3.125, 1.5625, 0.78125])

# Tableau contenant les valeurs d'absorbance correspondantes
absorbances = np.array([1.2, 0.6, 0.3, 0.15, 0.075, 0.0375, 0.01875, 0.009375])
```

### Étape 3 : Tracé de la courbe étalon
