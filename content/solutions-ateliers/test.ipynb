```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Charger les données des fichiers CSV
temperature_data = pd.read_csv('temperature_data.csv')
temperature_data_histo = pd.read_csv('temperature_data_histo.csv')
water_distribution_data = pd.read_csv('water_distribution_data.csv')
precipitation_height_data = pd.read_csv('precipitation_height_data.csv')

# 1. Analyse des températures mensuelles de Laval et Montréal pour l'année 2023
def analyze_temperature_data():
    # Grouper par mois et calculer la température moyenne pour chaque ville
    monthly_avg_temp = temperature_data.groupby('Month').mean()
    #monthly_avg_temp2 = temperature_data.groupby('Month').mean()
    
    # Tracer les données en graphique en ligne
    plt.figure(figsize=(10, 6))
    plt.plot(monthly_avg_temp.index, monthly_avg_temp['Laval_Temperature'], label='Laval')
    plt.plot(monthly_avg_temp.index, monthly_avg_temp['Montreal_Temperature'], label='Montréal')
    plt.xlabel('Mois')
    plt.ylabel('Température Moyenne (°C)')
    plt.title('Température Moyenne Mensuelle à Laval et Montréal (2023)')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Tracer les données en histogramme
    plt.figure(figsize=(10, 6))
    bar_width = 0.35
    index = range(len(monthly_avg_temp.index))

    plt.bar(index, monthly_avg_temp['Laval_Temperature'], bar_width, label='Laval')
    plt.bar([i + bar_width for i in index], monthly_avg_temp['Montreal_Temperature'], bar_width, label='Montréal')

    plt.xlabel('Mois')
    plt.ylabel('Température Moyenne (°C)')
    plt.title('Température Moyenne Mensuelle à Laval et Montréal (2023)')
    plt.xticks([i + bar_width / 2 for i in index], monthly_avg_temp.index)
    plt.legend()
    plt.grid(True)
    plt.show()


# 2. Analyse de la distribution de l'eau sur la Terre pour l'année 2023
def analyze_water_distribution():
    # Grouper par catégorie et calculer le pourcentage total pour chaque catégorie
    water_distribution = water_distribution_data.groupby('Category').sum()
    
    # Tracer les données
    plt.figure(figsize=(10, 6))
    plt.pie(water_distribution['Percentage'], labels=water_distribution.index, autopct='%1.1f%%', startangle=140)
    plt.title('Distribution de l\'Eau sur la Terre (2023)')
    plt.axis('equal')
    plt.show()

# 3. Analyse de la relation entre les précipitations (en mm) et la hauteur des arbres dans une forêt
def analyze_precipitation_height():
    # Tracer les données
    plt.figure(figsize=(10, 6))
    plt.scatter(precipitation_height_data['Precipitation (mm)'], precipitation_height_data['Height (m)'], alpha=0.5)
    plt.xlabel('Précipitations (mm)')
    plt.ylabel('Hauteur (m)')
    plt.title('Relation entre les Précipitations et la Hauteur des Arbres')
    plt.grid(True)
    plt.show()

# Fonction principale pour exécuter les analyses
def main():
    print("Analyse des températures mensuelles de Laval et Montréal pour l'année 2023")
    analyze_temperature_data()
    
    print("Analyse de la distribution de l'eau sur la Terre pour l'année 2023")
    analyze_water_distribution()
    
    print("Analyse de la relation entre les précipitations (en mm) et la hauteur des arbres dans une forêt")
    analyze_precipitation_height()

if __name__ == "__main__":
    main()
```
