```python
import pandas as pd

# Charger les données
df = pd.read_csv('data1.csv')

# Exploration des données
print(df.head())
print(df.info())
print(df.describe())

# Filtrer les données
df_filtered = df[df['Nombre'] > 10]

# ---

import matplotlib.pyplot as plt

# Histogramme
plt.hist(df['Nombre'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution des Valeurs')
plt.xlabel('Ecosysteme')
plt.ylabel('Espece')
plt.show()

# Appliquer les couleurs aux barres
couleurs = {
    'Forêt': 'green',
    'Désert': 'yellow',
    'Prairie': 'orange',
    'Marais': 'blue',
    'Océan': 'cyan'
}
couleurs_barres = [couleurs[ecosysteme] for ecosysteme in df['Ecosysteme']]

# Création du graphique en barres
plt.bar(df['Ecosysteme'], df['Nombre'], color=couleurs_barres, edgecolor='black')
plt.title('Distribution des Espèces par Écosystème')
plt.xlabel('Écosystème')
plt.ylabel('Nombre d\'Individus')
plt.show()

# Nuage de points
plt.scatter(df['Ecosysteme'], df['Espece'], color='red')
plt.title('Relation entre Espece et Ecosysteme')
plt.xlabel('Ecosysteme')
plt.ylabel('Espece')
plt.show()

#---

# Calcul de la moyenne
valeurs = df['Nombre'].tolist()

somme = 0
for valeur in valeurs:
    somme += valeur
moyenne = somme / len(valeurs)
print(f'Moyenne: {moyenne}')

#---

import numpy as np

# Régression linéaire
x = df['Ecosysteme'].values
y = df['Espece'].values
n = len(x)
xy = sum(x[i] * y[i] for i in range(n))
x_squared = sum(x[i] ** 2 for i in range(n))
a = (n * xy - sum(x) * sum(y)) / (n * x_squared - sum(x) ** 2)
b = (sum(y) - a * sum(x)) / n
print(f'Equation de la droite: y = {a}x + {b}')

# Visualisation de la régression linéaire
plt.scatter(x, y, color='blue')
plt.plot(x, a * x + b, color='red')
plt.title('Régression Linéaire')
plt.xlabel('Ecosysteme')
plt.ylabel('Espece')
plt.show()
```
