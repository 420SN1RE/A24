```python
import pandas as pd

# Charger les donnes
df = pd.read_csv('data1.csv')

# Exploration des donnes
print(df.head())
print(df.info())
print(df.describe())

# Filtrer les donnes
df_filtered = df[df['Valeur'] > 50]

#2. Visualisation
import matplotlib.pyplot as plt

# Histogramme
plt.hist(df['Valeur'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution des Valeurs')
plt.xlabel('Valeur')
plt.ylabel('Frquence')
plt.show()

# Nuage de points
plt.scatter(df['Principe'], df['Valeur'], color='red')
plt.title('Relation entre Principe et Valeur')
plt.xlabel('Principe')
plt.ylabel('Valeur')
plt.show()

#3. statistiques

# Calcul de la moyenne
valeurs = df['Valeur'].tolist()
somme = 0
for valeur in valeurs:
    somme += valeur
moyenne = somme / len(valeurs)
print(f'Moyenne: {moyenne}')

#4. Regression

import numpy as np

# Rgression linaire
x = df['Valeur'].values
y = df['Valeur'].values  # Utiliser une autre colonne pour y si ncessaire
n = len(x)
xy = sum(x[i] * y[i] for i in range(n))
x_squared = sum(x[i] ** 2 for i in range(n))
a = (n * xy - sum(x) * sum(y)) / (n * x_squared - sum(x) ** 2)
b = (sum(y) - a * sum(x)) / n
print(f'Equation de la droite: y = {a}x + {b}')

# Visualisation de la rgression linaire
plt.scatter(x, y, color='blue')
plt.plot(x, a * x + b, color='red')
plt.title('Rgression Linaire')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

```
