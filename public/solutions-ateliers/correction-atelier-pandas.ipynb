```python
# 7. Ajoutez une nouvelle colonne 'Index' à un dataframe qui est l'index actuel multiplié par 2.

df['index'] = df.index * 2
df
```

```python
# 8. Filtrez les lignes du dataframe où 'sepal.length' est supérieur à 5.

df_filtre = df[df['long_sepale'] > 5]
df_filtre
```

```python
# 9. Renommez la colonne 'variete' en 'especes'.

df.rename(columns={'variete': 'especes'}, inplace=True)
print(df)
```

```python
# 10. Groupez les données en fonction de la colonne 'especes' et calculez la moyenne de la colonne 'long_sepale' pour l'espèce 'setosa'.

#1. On **filtre** notre jeu de données sur la colonne especes
df_especes = df[df['especes'] == 'Setosa']
df_especes

#2. On **calcule la moyenne** des notes à l'aide d'une boucle `for`
somme_long_sepale = 0
for long_sepale in df_especes['long_sepale']:
   somme_long_sepale = somme_long_sepale + long_sepale

moyenne = somme_long_sepale / len(df_especes)
print(f"La moyenne des long_sepale par espèce est de {round(moyenne,2)}%")

```

```python
# 12. À l'aide d'une liste et de boucles, modifie les valeurs manquantes (None, NaN) par la valeur 0

# Convertir le DataFrame en liste
liste = df.values.tolist()

# Remplacer les valeurs manquantes par 0
for ligne in liste:
    for i in range(len(ligne)):
        if ligne[i] != ligne[i]:  # Vérifie si la valeur est NaN
            ligne[i] = 0

print(liste)
```
