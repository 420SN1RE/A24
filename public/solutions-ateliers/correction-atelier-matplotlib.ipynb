```python
## Exercice 3 :

import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("ventes.csv")
liste_mois  = df ['num_mois'].tolist()
ventes_creme_visage   = df ['creme_visage'].tolist()
ventes_creme_lave_visage  = df ['lave_visage'].tolist()
ventes_dentifrice = df ['dentifrice'].tolist()
ventes_savon_bain   = df ['savon_bain'].tolist()
ventes_shampoing   = df ['shampoing'].tolist()
ventes_creme_hydratante = df ['creme_hydratante'].tolist()

plt.plot(liste_mois, ventes_creme_visage,   label = 'Ventes crème visage', marker='o', linewidth=3)
plt.plot(liste_mois, ventes_creme_lave_visage,   label = 'Ventes crème lave visage',  marker='o', linewidth=3)
plt.plot(liste_mois, ventes_dentifrice, label = 'Ventes dentifrice', marker='o', linewidth=3)
plt.plot(liste_mois, ventes_savon_bain, label = 'Ventes savon de bain', marker='o', linewidth=3)
plt.plot(liste_mois, ventes_shampoing, label = 'Ventes shampoing', marker='o', linewidth=3)
plt.plot(liste_mois, ventes_creme_hydratante, label = 'Ventes crème hydratante', marker='o', linewidth=3)

plt.xlabel('Mois (1-12)')
plt.ylabel('Unités de ventes')
plt.legend(loc='upper left')
plt.xticks(liste_mois)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Ventes')
plt.savefig('exercice3.png', dpi=150)
plt.show()
```

```python
## Exercice 8 :

import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("ventes.csv")
liste_mois  = df ['num_mois'].tolist()

labels = ['crème visage', 'lave visage', 'dentifrice', 'savon bain', 'shampoing', 'hydratant']
ventes   = [df ['creme_visage'].sum(), df ['lave_visage'].sum(), df ['dentifrice'].sum(), 
         df ['savon_bain'].sum(), df ['shampoing'].sum(), df ['creme_hydratante'].sum()]
plt.axis("equal")
plt.pie(ventes, labels=labels, autopct='%1.1f%%')
plt.legend(loc='lower right')
plt.title('Ventes')
plt.savefig('exercice8.png', dpi=150)
plt.show()
```

```python
## Exercice 10 :

import pandas as pd
import matplotlib.pyplot as plt  

df = pd.read_csv("ventes.csv")
liste_mois  = df ['num_mois'].tolist()

ventes_creme_visage   = df ['creme_visage'].tolist()
ventes_creme_lave_visage  = df ['lave_visage'].tolist()
ventes_dentifrice = df ['dentifrice'].tolist()
ventes_savon_bain   = df ['savon_bain'].tolist()
ventes_shampoing   = df ['shampoing'].tolist()
ventes_creme_hydratante = df ['creme_hydratante'].tolist()

plt.plot([],[],color='m', label='Crème visage', linewidth=5)
plt.plot([],[],color='c', label='Lave visage', linewidth=5)
plt.plot([],[],color='r', label='Dentifrice', linewidth=5)
plt.plot([],[],color='k', label='Savon bain', linewidth=5)
plt.plot([],[],color='g', label='Shampoing', linewidth=5)
plt.plot([],[],color='y', label='Crème hydratante', linewidth=5)

plt.stackplot(liste_mois, ventes_creme_visage, ventes_creme_lave_visage, ventes_dentifrice, 
              ventes_savon_bain, ventes_shampoing, ventes_creme_hydratante, 
              colors=['m','c','r','k','g','y'])

plt.xlabel('Mois (1-12)')
plt.ylabel('Unités de ventes')
plt.title('Toutes les données de vente de produits en utilisant un graphique empilé')
plt.legend(loc='upper left')
plt.savefig('exercice10.png', dpi=150)
plt.show()

```
